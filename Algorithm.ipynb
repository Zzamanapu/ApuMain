```python
# dynamic sliding window practice
# minimum sum subarray
import math
arr = [1, 2, 4, 5, 9, 7]
s = 5
ans = math.inf
sum = 0
L = 0

for R in range(len(arr)):
  sum += arr[R]
  while sum >= s:
    ans = min(ans, R - L + 1)
    sum -= arr[L]
    L += 1

print(ans)
```

```python
# [Naive Approach] By Using Array - O((n+m) Ã— log(n+m)) Time and O(n+m) Space

# class Node:
#   def __init__(self, x):
#     self.data = x
#     self.next = None

# def sortedMerge(head1, head2):
#   arr = []

#   while head1 is not None:
#     arr.append(head1.data)
#     head1 = head1.next

#   while head2 is not None:
#     arr.append(head2.data)
#     head2 = head2.next

#   arr.sort()

#   dummy = Node(-1)
#   curr = dummy

#   for val in arr:
#     curr.next = Node(val)
#     curr = curr.next

#   return dummy.next

# def printList(node):
#   while node is not None:
#     print(f"{node.data}", end="")
#     if node.next is not None:
#       print(" -> ", end = "")
#     node = node.next
#   print()

# if __name__ == "__main__":
#   head1 = Node(5)
#   head1.next = Node(10)
#   head1.next.next = Node(15)
#   head1.next.next.next = Node(40)

#   head2 = Node(2)
#   head2.next = Node(3)
#   head2.next.next = Node(20)

#   res = sortedMerge(head1, head2)

#   printList(res)





#[Better Approach] Using Recursive Merge - O(n+m) Time and O(n+m) Space

# class Node:
#   def __init__(self, x):
#     self.data = x
#     self.next = None

# def sortedMerge(head1, head2):
#   if head1 is None:
#     return head1
#   if head2 is None:
#     return head2

#   if head1.data <= head2.data:
#     head1.next = sortedMerge(head1.next, head2)
#     return head1
#   else:
#     head2.next = sortedMerge(head1, head2.next)
#     return head2

# def printList(node):
#   while node is not None:
#     print(f"{node.data}", end="")
#     if node.next is not None:
#       print(" -> ", end = "")
#     node = node.next
#   print()

# if __name__ == "__main__":
#   head1 = Node(5)
#   head1.next = Node(10)
#   head1.next.next = Node(15)
#   head1.next.next.next = Node(40)

#   head2 = Node(2)
#   head2.next = Node(3)
#   head2.next.next = Node(20)

#   res = sortedMerge(head1, head2)

#   printList(res)





#[Efficient Approach] Using Iterative Merge - O(n+m) Time and O(1) Space

class Node:
  def __init__(self, x):
    self.data = x
    self.next = None

def sortedMerge(head1, head2):
  dummy = Node(-1)
  curr = dummy

  while head1 is not None and head2 is not None:
    if head1.data <= head2.data:
      curr.next = head1
      head1 = head1.next
    else:
      curr.next = head2
      head2 = head2.next
    curr = curr.next


  if head1 is not None:
    curr.next = head1
  if head2 is not None:
    curr.next = head2

  return dummy.next

def printList(node):
  while node is not None:
    print(f"{node.data}", end="")
    if node.next is not None:
      print(" -> ", end = "")
    node = node.next
  print()

if __name__ == "__main__":
  head1 = Node(5)
  head1.next = Node(10)
  head1.next.next = Node(15)
  head1.next.next.next = Node(40)

  head2 = Node(2)
  head2.next = Node(3)
  head2.next.next = Node(20)

  res = sortedMerge(head1, head2)

  printList(res)

```
